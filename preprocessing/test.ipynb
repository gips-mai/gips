```python

def extract_image_response(nested_dict):
    """
    Extracts the image file and the generated response from the dict.

    Parameter:
    nested_dict (dict): a dicitionary containing the keys 'image_file' and 'response'

    RÃ¼ckgabewerte:
    tuple: A tuple consisting of the image file and the generated response.
    """
    return nested_dict['image_file'], nested_dict['response']
```

```python

def process_csv_directory(directory):
    """
    Processes all CSV files in a directory, extracts nested dictionaries and creates a DataFrame where each key is mapped as a column.

    Parameters:
    directory (str): The path to the directory containing the CSV files.

    Return values:
    pd.DataFrame: A DataFrame with the extracted 'image_file' and 'response' data.
    """

    columns = ['image_file', 'response']
    extracted_data = []

    for file_name in os.listdir(directory):
        if file_name.endswith(".csv"): 
            file_path = os.path.join(directory, file_name)
            df = pd.read_csv(file_path)  

            
            for col in df.columns:
                for item in df[col].dropna():
                    nested_dict = eval(item)  # Convert the string into a dictionary
                    extracted_data.append(extract_image_response(nested_dict))

 
    extracted_df = pd.DataFrame(extracted_data, columns=columns)
    return extracted_df
```
